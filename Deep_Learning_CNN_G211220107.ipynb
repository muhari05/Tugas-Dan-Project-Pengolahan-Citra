```python

# 1. IMPORT LIBRARY

import tensorflow as tf
from tensorflow.keras import layers, models
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report, accuracy_score, precision_score, recall_score, f1_score
import pandas as pd



# 2. LOAD & PREPROCESSING DATASET

def load_dataset(dataset_choice):
    print(f"\n Memuat Dataset: {dataset_choice}")

    if dataset_choice == "fashion_mnist":
        (x_train, y_train), (x_test, y_test) = tf.keras.datasets.fashion_mnist.load_data()

        x_train = x_train / 255.0
        x_test  = x_test / 255.0

        x_train = x_train[..., None]
        x_test  = x_test[..., None]

        input_shape = (28, 28, 1)
        dataset_name = "Fashion-MNIST"

    elif dataset_choice == "cifar10":
        (x_train, y_train), (x_test, y_test) = tf.keras.datasets.cifar10.load_data()

        x_train = x_train / 255.0
        x_test  = x_test / 255.0

        y_train = y_train.flatten()
        y_test  = y_test.flatten()

        input_shape = (32, 32, 3)
        dataset_name = "CIFAR-10"

    print("Shape data training:", x_train.shape)
    print("Shape data testing :", x_test.shape)

    return x_train, y_train, x_test, y_test, input_shape, dataset_name



# 3. MEMBANGUN MODEL CNN

def build_cnn(input_shape, num_classes):

    model = models.Sequential([
        layers.Conv2D(32, (3,3), activation='relu', input_shape=input_shape),
        layers.MaxPooling2D((2,2)),
        layers.Conv2D(64, (3,3), activation='relu'),
        layers.MaxPooling2D((2,2)),
        layers.Flatten(),
        layers.Dense(128, activation='relu'),
        layers.Dense(num_classes, activation='softmax')
    ])

    model.compile(
        optimizer='adam',
        loss='sparse_categorical_crossentropy',
        metrics=['accuracy']
    )


    # SUMMARY MODEL

    print("\n SUMMARY ARSITEKTUR MODEL CNN")
    model.summary()


    # TABEL ARSITEKTUR CNN

    summary_data = []
    for layer in model.layers:
        try:
            output_shape = tuple(layer.output.shape)
        except:
            output_shape = "Tidak tersedia"

        summary_data.append({
            "Layer Name": layer.name,
            "Layer Type": layer.__class__.__name__,
            "Output Shape": output_shape,
            "Jumlah Parameter": layer.count_params()
        })

    summary_df = pd.DataFrame(summary_data)
    print("\n TABEL ARSITEKTUR CNN")
    print(summary_df)

    return model


# 4. EVALUASI MODEL

def evaluate_model(model, x_test, y_test, dataset_name):

    y_pred = np.argmax(model.predict(x_test), axis=1)

    acc  = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred, average='weighted')
    rec  = recall_score(y_test, y_pred, average='weighted')
    f1   = f1_score(y_test, y_pred, average='weighted')

    print(f"\n Evaluasi Model ({dataset_name})")
    print(f"Akurasi  : {acc:.4f}")
    print(f"Presisi  : {prec:.4f}")
    print(f"Recall   : {rec:.4f}")
    print(f"F1 Score : {f1:.4f}")

    print("\nClassification Report:")
    print(classification_report(y_test, y_pred))

    plt.figure(figsize=(8, 8))
    for i in range(9):
        plt.subplot(3, 3, i+1)
        if x_test.shape[-1] == 1:
            plt.imshow(x_test[i].reshape(28,28), cmap='gray')
        else:
            plt.imshow(x_test[i])
        plt.title(f"Pred: {y_pred[i]}, True: {y_test[i]}")
        plt.axis('off')

    plt.suptitle(f"Visualisasi Hasil Prediksi - {dataset_name}")
    plt.show()

    return acc, prec, rec, f1



# 5. VISUALISASI HASIL (AKURASI & LOSS)

def plot_history(history, dataset_name):
    plt.figure(figsize=(12, 4))

    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label='Train Acc')
    plt.plot(history.history['val_accuracy'], label='Val Acc')
    plt.title(f'Akurasi Model ({dataset_name})')
    plt.xlabel('Epoch')
    plt.ylabel('Akurasi')
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='Train Loss')
    plt.plot(history.history['val_loss'], label='Val Loss')
    plt.title(f'Loss Model ({dataset_name})')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()

    plt.show()



# 6. KOMPILASI & TRAINING (FASHION-MNIST)
x_train_f, y_train_f, x_test_f, y_test_f, shape_f, name_f = load_dataset("fashion_mnist")
model_f = build_cnn(shape_f, 10)

history_f = model_f.fit(
    x_train_f, y_train_f,
    epochs=5,
    batch_size=128,
    validation_data=(x_test_f, y_test_f),
    verbose=1
)

plot_history(history_f, name_f)
acc_f, prec_f, rec_f, f1_f = evaluate_model(model_f, x_test_f, y_test_f, name_f)


# 7. KOMPILASI & TRAINING (CIFAR-10)

x_train_c, y_train_c, x_test_c, y_test_c, shape_c, name_c = load_dataset("cifar10")
model_c = build_cnn(shape_c, 10)

history_c = model_c.fit(
    x_train_c, y_train_c,
    epochs=5,
    batch_size=128,
    validation_data=(x_test_c, y_test_c),
    verbose=1
)

plot_history(history_c, name_c)
acc_c, prec_c, rec_c, f1_c = evaluate_model(model_c, x_test_c, y_test_c, name_c)



# 8. PERBANDINGAN HASIL EVALUASI MODEL

comparison_df = pd.DataFrame([
    {"Dataset": name_f, "Accuracy": acc_f, "Precision": prec_f, "Recall": rec_f, "F1 Score": f1_f},
    {"Dataset": name_c, "Accuracy": acc_c, "Precision": prec_c, "Recall": rec_c, "F1 Score": f1_c}
])

print("\n PERBANDINGAN METRIK CNN")
print(comparison_df)



# 9. KESIMPULAN HASIL PERCOBAAN

print("""
KESIMPULAN:
Model CNN menunjukkan performa lebih baik pada dataset Fashion-MNIST
yang memiliki kompleksitas lebih rendah dibanding CIFAR-10.
Dataset CIFAR-10 membutuhkan arsitektur CNN yang lebih dalam
untuk memperoleh performa optimal.
""")
```
